## Feature Engineering üõ†Ô∏è

In this section, the goal is to  create two new columns:

1. **Total Score:** We'll calculate the total score for each student by summing their math, reading, and writing scores.

2. **Average Score:** To understand the overall performance, we'll compute the average score for each student, considering their math, reading, and writing scores.

These new features will provide a holistic view of student performance and help us uncover meaningful patterns. Let's harness the power of feature engineering to unlock more knowledge from our dataset! üõ†Ô∏èüìäüéì


````python
df['total_score'] = df['math_score'] + df['reading_score'] + df['writing_score']
df['average'] = df['total_score'] / 3
````

````python
reading_full = df[df['reading_score'] == 100]['average'].count()
writing_full = df[df['writing_score'] == 100]['average'].count()
math_full = df[df['math_score'] == 100]['average'].count()

print(f'The number of studnts with full marks in Mathematics: {math_full}')
print(f'The number of studnts with complete marks in Writing: {writing_full}')
print(f'The number of studnts with complete marks in Reading: {reading_full}')
````

````python
reading_less_20 = df[df['reading_score']<=20]['average'].count()
writing_less_20 = df[df['writing_score']<=20]['average'].count()
math_less_20 = df[df['math_score'] <= 20]['average'].count()

print(f'The number of student who score less than 20 marks in Mathematics was: {math_less_20}')
print(f'The number of student who score less than 20 marks in writing was: {writing_less_20}')
print(f'The number of student who score less than 20 marks in Reading was: {reading_less_20}')
````
